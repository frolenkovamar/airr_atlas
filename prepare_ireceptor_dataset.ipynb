```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Oslo/data/ireceptor/3M.tsv', sep='\t', dtype={'sequence_id': 'str',	'sequence_aa': 'str',	'productive': 'str',	'locus	v_call': 'str',	'd_call': 'str',	'j_call': 'str',	'c_call': 'str',	'junction_aa': 'str',	'duplicate_count': 'float',	'repertoire_id': 'str',	'cdr1_aa': 'str',	'cdr2_aa': 'str',	'cdr3_aa': 'str'})
df
```

```python
columns_to_check = ['cdr3_aa', 'v_call', 'd_call', 'j_call', 'c_call', 'junction_aa', 'repertoire_id', 'sequence_id']
df = df.dropna(subset=columns_to_check)
df = df.drop_duplicates(subset='junction_aa', keep='first')

df
```

```python
# Select 800,000 random rows
sampled_df = df.sample(n=800000, random_state=42)  # Use a specific random_state for reproducibility
sampled_df['id'] = range(1, len(sampled_df) + 1)

# Save the sampled DataFrame as a CSV
sampled_df.to_csv('/content/drive/MyDrive/Oslo/data/ireceptor/H_CDR3.csv', index=False, sep=';')
```

```python
sampled_df
```

```python
fasta_file = '/content/drive/MyDrive/Oslo/data/ireceptor/H_CDR3.fa'

with open(fasta_file, 'w') as file:
    for index, row in sampled_df.iterrows():
        id = row['id']
        cdr_aa = row['junction_aa']

        # Write the FASTA entry
        file.write(f'>{id}\n{cdr_aa}\n')
```

# **Trastuzumab merge**

```python
import pandas as pd


path = '/content/drive/MyDrive/Oslo/data/trastuzumab/'

# Read each CSV file
df_lb = pd.read_csv(path + 'Unique_lb.csv')
df_hb = pd.read_csv(path + 'Unique_hb.csv')
df_mb = pd.read_csv(path + 'Unique_mb.csv')

# Add 'affinity' column with respective labels
df_lb['affinity'] = 'lb'
df_hb['affinity'] = 'hb'
df_mb['affinity'] = 'mb'

# Concatenate the dataframes
result = pd.concat([df_lb, df_hb, df_mb], ignore_index=True)

# Add 'id' column starting from 1
result['id'] = range(1, len(result) + 1)

result['aa_cdr3_seq'] = 'CSR' + result['aa_cdr3_seq'] + 'YW'

result
```

```python
# Save the merged dataframe to a new CSV file
result.to_csv(path + 'Merged_unique.csv', index=False)

# Save fasta
with open(path + 'Merged_unique.fa', 'w') as file:
    for index, row in result.iterrows():
        id = row['id']
        cdr_aa = row['aa_cdr3_seq']

        # Write the FASTA entry
        file.write(f'>{id}\n{cdr_aa}\n')
```

# **Covidab**

```python
path = '/content/drive/MyDrive/Oslo/data/covabdab/'

df = pd.read_csv(path + 'CoV-AbDab_130623.csv', sep=',')
df = df.dropna(subset='CDRH3')

# Add 'id' column starting from 1
df['id'] = range(1, len(df) + 1)

df['CDRH3'] = 'C' + df['CDRH3'] + 'W'

df.head()
```

```python
df.columns
```

```python
# Count the occurrences of each unique value in the 'Protein + Epitope' column
occurrences = df['Protein + Epitope'].value_counts()

# Display the occurrences of each unique value
print("Occurrences of each unique value in 'Protein + Epitope' column:")
print(occurrences)
```

```python
# Save the merged dataframe to a new CSV file
df.to_csv(path + 'covabdab.csv', index=False)

# Save fasta
with open(path + 'covabdab.fa', 'w') as file:
    for index, row in df.iterrows():
        id = row['id']
        cdr_aa = row['CDRH3']

        # Write the FASTA entry
        file.write(f'>{id}\n{cdr_aa}\n')
```
